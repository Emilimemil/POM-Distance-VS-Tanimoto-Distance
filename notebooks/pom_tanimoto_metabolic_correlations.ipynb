```python
import os
import pandas as pd
import plotly.express as px
from scipy.stats import pearsonr


```

```python
# 1. Load the dataset
#
# To reproduce this analysis:
# 1. Download metabolite_distance.csv from the original Qian et al. (2023) dataset.
# 2. Place the file in:
#       data/metabolite_distance.csv
#    when running locally from the GitHub repository.
# 3. Alternatively, upload the CSV directly when using Google Colab.
#
# The code below automatically detects common file locations for both
# local Jupyter environments and Google Colab.

possible_paths = [
    "../data/metabolite_distance.csv",   # local GitHub/Jupyter
    "metabolite_distance.csv",           # normal Colab upload
    "/metabolite_distance.csv"           # current Colab session
]

for path in possible_paths:
    if os.path.exists(path):
        print(f"Loading: {path}")
        df = pd.read_csv(path)
        break
else:
    raise FileNotFoundError("metabolite_distance.csv not found")
```

```python
print(df.columns)
```

```python
# 2. Create an interactive scatter plot
r, _ = pearsonr(df["POM distance"], df["tanimoto distance"])

fig = px.scatter(
    df,
    x="metabolic distance",
    y="POM distance",
    # 'hover_name' controls the main label in the tooltip
    hover_name="metabolite1",
    # 'hover_data' controls extra fields you want to see on hover
    hover_data=["specie","metabolite1","metabolite2", "smiles1", "smiles2"],
    title="Metabolic Distance vs. POM Distance",
    trendline="ols",
    trendline_color_override="green"
)

fig.show()
```

```python
# 2. Create an interactive scatter plot
fig = px.scatter(
    df,
    x="metabolic distance",
    y="tanimoto distance",
    # 'hover_name' controls the main label in the tooltip
    hover_name="metabolite1",
    # 'hover_data' controls extra fields you want to see on hover
    hover_data=["specie","metabolite1","metabolite2", "smiles1", "smiles2"],
    title="tanimoto distance vs. metabolic Distance",
    trendline="ols",
    trendline_color_override="green"
)
r, _ = pearsonr(df["metabolic distance"], df["tanimoto distance"])
r_text = f"Pearson r = {r:.2f}"

fig.add_annotation(
    text=r_text,
    xref="paper", yref="paper",
    x=0.05, y=0.95,  # position in top-left corner
    showarrow=False,
    font=dict(size=14)
)


fig.show()
```
