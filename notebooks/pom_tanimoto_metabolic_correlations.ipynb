```python
import pandas as pd
import plotly.express as px
from scipy.stats import pearsonr


```

```python
# 1. Load your data
# To run this notebook, download metabolite_distance.csv from the original Qian et al. dataset and place it in:
# data/metabolite_distance.csv
df = pd.read_csv("data/metabolite_distance.csv")
```

```python
print(df.columns)
```

```python
# 2. Create an interactive scatter plot
r, _ = pearsonr(df["POM distance"], df["tanimoto distance"])

fig = px.scatter(
    df,
    x="metabolic distance",
    y="POM distance",
    # 'hover_name' controls the main label in the tooltip
    hover_name="metabolite1",
    # 'hover_data' controls extra fields you want to see on hover
    hover_data=["specie","metabolite1","metabolite2", "smiles1", "smiles2"],
    title="Metabolic Distance vs. POM Distance",
    trendline="ols",
    trendline_color_override="green"
)

fig.show()
```

```python
# 2. Create an interactive scatter plot
fig = px.scatter(
    df,
    x="metabolic distance",
    y="tanimoto distance",
    # 'hover_name' controls the main label in the tooltip
    hover_name="metabolite1",
    # 'hover_data' controls extra fields you want to see on hover
    hover_data=["specie","metabolite1","metabolite2", "smiles1", "smiles2"],
    title="tanimoto distance vs. metabolic Distance",
    trendline="ols",
    trendline_color_override="green"
)
r, _ = pearsonr(df["metabolic distance"], df["tanimoto distance"])
r_text = f"Pearson r = {r:.2f}"

fig.add_annotation(
    text=r_text,
    xref="paper", yref="paper",
    x=0.05, y=0.95,  # position in top-left corner
    showarrow=False,
    font=dict(size=14)
)


fig.show()
```

```python
filtered_df = df[
    (df["POM distance"] < 0.2) &
    (df["tanimoto distance"] > 0.6)
]
columns_needed = ["specie", "metabolite1", "metabolite2",
                  "tanimoto distance", "POM distance", "metabolic distance"]

# If any of these columns are missing or named differently, you’ll get a KeyError—adjust as needed
df_view = filtered_df[columns_needed]

print(df_view)
```

```python
import webbrowser

base_url = "https://metacyc.org/META/substring-search?type=NIL&object="

for idx, row in df_view.iterrows():
    # For each column we want to search:
    for col in ["specie", "metabolite1", "metabolite2"]:
        search_term = row[col]
        if pd.notna(search_term):   # skip NaN or empty
            search_url = base_url + str(search_term)
            # This opens the URL in your default browser.
            # NOTE: This could open many tabs if you have 10 rows x 3 columns = 30 searches.
            webbrowser.open_new_tab(search_url)
```
